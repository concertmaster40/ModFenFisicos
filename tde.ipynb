1. Escreva uma função em Python que compute o valor da função abaixo, emitindo uma mensagem de erro caso  a  função  receba um  valor não  permitido.  Além  disso,  a função  deverá  plotar  um gráfico da função no intervalo entre ±10, no domínio dos números reais.

 $f(x) = \frac{1}{\sqrt{2 - x}}$

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def f(x):
  if x < 2:
    return (1 / math.sqrt(2 - x))
  else:
    np.nan

def plot_graph():
  eixo_x  = np.linspace(-10, 11, 100)
  eixo_y = [f(x) for x in eixo_x]


  plt.plot(eixo_x, eixo_y)
  plt.xlabel("x")
  plt.ylabel('$f(x)$')
  plt.title(r'$f(x) = \frac{1}{\sqrt{2-x}}$')
  plt.grid()
  plt.show()


x = int(input("Digite um valor para x: "))
if x > 2:
  print("Valor inválido") # Mensagem de erro caso a função receba um valor não permitido
else:
  y = f(x)
  print(f"f({x}) = {y}\nAbaixo está o gráfico da função:")
  plot_graph()
```

2. Utilizando a linguagem Python, e somente as bibliotecas Matplotlib e Numpy, crie uma função
para plotar uma raquete de tênis de mesa. No seu gráfico a raquete deve ser plotada de forma
que seja possível observar tanto o cabo quanto a área de rebatimento.


```python
#Valores x para elipse
x = np.linspace(-4, 4, 400)
#Valores y para elipse
y_positivo = 5 * np.sqrt(1 - (x**2 / 16))
y_negativo = -5 * np.sqrt(1 - (x**2 / 16))

def plot_elipse_segmentos(y_positivo, y_negativo, x1, y1, x2, y2, x3, y3, x4, y4):
    # Elipse
    plt.plot(x, y_positivo)
    plt.plot(x, y_negativo)
    plt.plot(x, -y_positivo)
    plt.plot(x, -y_negativo)

    # Segmentos das Retas
    plt.plot([x1, x2, x3, x4], [y1, y2, y3, y4])

# Config plotagem
plt.figure(figsize=(8, 6))
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Raquete de tênis')
plt.grid(True)


# Plotar elipse e segmentos
plot_elipse_segmentos(y_positivo, y_negativo, -1, -4.9, -1, -10, 1, -10, 1, -4.9)

plt.show()
```

3. Utilizando a linguagem Python, nenhuma biblioteca extra e as técnicas de List Comprehension
crie uma função que devolva o resultado da fórmula a seguir para 𝑛 elementos.

$$f(x) = 4\left(1-\sum_{1}^{n} \frac{1}{1 + 2n}\right)$$

```python
from IPython.display import display, Math

# Função cálculo da somatória
def somatoria(n):
    somatoria = [(1 / (1 + 2 * n)) for n in range(1, n+1)] #List Comprehension

    somas_totais = 0
    for i in range(len(somatoria)):
        somas_totais += somatoria[i]
    return somas_totais

# Função da operação completa com a somatória
def ops(n):
    operacao = 4 * (1 - somatoria(n))
    return operacao

n = int(input("Digite o valor para n: "))
print("\n")
resultado = ops(n)

# Definir a expressão da soma usando LaTeX
soma_expression = r"4\left(1-\sum_{n=1}^{" + str(n) + r"} \frac{1}{" + r"1 + 2n}\right)\approx" + str("{:.2f}".format(resultado))


# Exibir a expressão da soma usando LaTeX
display(Math(soma_expression))
```

4. Usando a linguagem Python, sem utilizar as bibliotecas Sympy e Numpy, escreva uma função
que receba um valor 𝑛 no domínio dos números reais e devolva o limite da função a seguir
quando 𝑥 tende a 𝑛.

 $f(x) = \frac{3x^2 - 1}{\sqrt{x}}$

```python
import matplotlib.pyplot as plt
import sympy as sp
import numpy as np
from IPython.display import display, Math

def plotar_funcao():
    # Definindo a variável simbólica
    x = sp.Symbol('x')

    # Definindo a função
    y = (3 * x ** 2 - 1) / sp.sqrt(x)

    # Plotando o gráfico da função usando matplotlib
    x_vals = np.linspace(0.01, 2, 400)  # Começar por 0.01 para evitar divisão por 0
    y_vals = [(2 * x_val ** 2 - 1) / (np.sqrt(x_val)) for x_val in x_vals]

    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y_vals, label=r'$f(x) = \frac{3x^2 - 1}{\sqrt{x}}$')
    plt.title('Gráfico da função $f(x) = \\frac{3x^2 - 1}{\sqrt{x}}$')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.show()

    # Obter o valor de n do usuário
    n = float(input("Digite o valor de n para calcular o limite: "))

    # Calcular o limite da função
    limite = sp.limit(y, x, n)
    print("\n")
    display(Math(r"\lim_{{x \to {}}} \left( {} \right) = {}".format(int(n), sp.latex(y), sp.latex(limite))))

plotar_funcao()
```
