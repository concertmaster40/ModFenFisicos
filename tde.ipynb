1. Escreva uma fun√ß√£o em Python que compute o valor da fun√ß√£o abaixo, emitindo uma mensagem de erro caso  a  fun√ß√£o  receba um  valor n√£o  permitido.  Al√©m  disso,  a fun√ß√£o  dever√°  plotar  um gr√°fico da fun√ß√£o no intervalo entre ¬±10, no dom√≠nio dos n√∫meros reais. ùëì(ùë•)=1/‚àö2‚àíùë•

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def f(x):
  if x < 2:
    return (1 / math.sqrt(2 - x))
  else:
    np.nan

def plot_graph():
  eixo_x  = np.linspace(-10, 11, 100)
  eixo_y = [f(x) for x in eixo_x]


  plt.plot(eixo_x, eixo_y)
  plt.xlabel("x")
  plt.ylabel('$f(x)$')
  plt.title(r'$f(x) = \frac{1}{\sqrt{2-x}}$')
  plt.grid()
  plt.show()


x = int(input("Digite um valor para x: "))
if x > 2:
  print("Valor inv√°lido")
else:
  y = f(x)
  print(f"f({x}) = {y}\n Abaixo est√° o gr√°fico da fun√ß√£o:")
  plot_graph()
```

2. Utilizando a linguagem Python, e somente as bibliotecas Matplotlib e Numpy, crie uma fun√ß√£o
para plotar uma raquete de t√™nis de mesa. No seu gr√°fico a raquete deve ser plotada de forma
que seja poss√≠vel observar tanto o cabo quanto a √°rea de rebatimento.


```python
#Valores x para elipse
x = np.linspace(-4, 4, 400)
#Valores y para elipse
y_positivo = 5 * np.sqrt(1 - (x**2 / 16))
y_negativo = -5 * np.sqrt(1 - (x**2 / 16))

def plot_elipse_segmentos(y_positivo, y_negativo, x1, y1, x2, y2, x3, y3, x4, y4):
    #Elipse
    plt.plot(x, y_positivo)
    plt.plot(x, y_negativo)
    plt.plot(x, -y_positivo)
    plt.plot(x, -y_negativo)

    #Segmentos das Retas
    plt.plot([x1, x2, x3, x4], [y1, y2, y3, y4])

#Config plotagem
plt.figure(figsize=(8, 6))
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Raquete de t√™nis')
plt.grid(True)


#Plotar eclipse e segmentos
plot_elipse_segmentos(y_positivo, y_negativo, -1, -4.9, -1, -10, 1, -10, 1, -4.9)

plt.show()
```

```python
#Fun√ß√£o c√°lculo da somatoria
def somatoria(n):
  somatoria = [(1 / (1 + 2 * n)) for n in range(1, n+1)]

  somas_totais = 0
  for i in range(len(somatoria)):
    somas_totais += somatoria[i]
  return somas_totais

#Fun√ß√£o da opera√ß√£o completa com a somatoria
def ops(n):
  operacao = 4 * (1 - somatoria(n))
  return operacao

n = int(input("Digite o valor para n: "))
print(f'Resultado da somat√≥ria = {ops(n)}')
```
