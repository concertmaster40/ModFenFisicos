Resolva os seguintes limites algebricamente: (a) para as funções
tendendo ao infinito; (b) para as funções tendendo a zero. Depois
verifique os limites encontrados com o sympy e trace os gráficos das
funções.

$f(x) = \frac{x - 3}{7 - x}$


```python
import matplotlib.pyplot as plt
import math
import sympy as sp
import numpy as np
from IPython.display import display, Math


def plotar_funcao():
    # Definindo a variável simbólica
    x = sp.Symbol('x')

    # Definindo a função
    y = (x - 3) / (7 - x)

    # Plotando o gráfico da função usando sympy
    #sp.plot(y, (x, -2, 2), title=r'Gráfico da função $f(x) = \frac{x - 3}{7 - x}$')

    # Plotando o gráfico da função usando matplotlib
    x_vals = np.linspace(-2, 2, 400)
    y_vals = [(x_val - 3) / (7 - x_val) for x_val in x_vals]

    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y_vals, label=r'$f(x) = \frac{x - 3}{7 - x}$')
    plt.title('Gráfico da função $f(x) = \\frac{x - 3}{7 - x}$')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
    plt.legend()
    plt.show()

    # Calcular os limites da função
    limite_infinito = sp.limit(y, x, sp.oo)
    print("\n")
    display(Math(r"\lim_{{x \to \infty}} \left( {} \right) = {}".format(sp.latex(y), sp.latex(limite_infinito))))
    limite_zero = sp.limit(y, x, 0)
    print("\n")
    display(Math(r"\lim_{{x \to 0}} \left( {} \right) = {}".format(sp.latex(y), sp.latex(limite_zero))))

plotar_funcao()
```
